**Funcion basica**

**Con parametros por defecto**

```python
def potencia(base, exponente=2):
  '''Eleva base a exponente. Por defecto calcula el cuadrado.'''
  return base ** exponente

print(potencia(5))
print(potencia(2,3))
```

```python
def min_max(lista):
  return min(lista), max(lista)

minimo, maximo = min_max([3,1,7,2,5])

print("Minimo: ",minimo,"\nMaximo: ",maximo)
```

**Vista vs. Copia**

```python
vista = A[0, :] # es una vista
copia = A[0, :].copy() # es una copia independiente
vista[0] = 999 # modifica A también!
copia[0] = 888 # NO modifica A
print(A[0, :]) #vista modificó el original
print(np.shares_memory(A, vista)) # True
print(np.shares_memory(A, copia)) # False
```

***Indexación Booleana***

***Estadística y Reducciones***

```python
M = np.array([[4, 7, 2, 8],
[3, 11, 5, 1],
[6, 9, 0, 4]])
```

```python
print(M.sum())
print(M.mean())
print(M.max())
print(M.min())
```

**Reducción por eje**

```python
print("suma por columna: ",M.sum(axis=0))
print("suma por fila: ",M.sum(axis=1))
print("media de cada columna: ",M.mean(axis=0))
print("maximo de cada fila: ",M.max(axis=1))
```

**argmin / argmax — índice del extremo**

```python
print(M.argmin(axis=1))
print(M.argmax(axis=0))
```

**std y var**

```python
print(M.std()) # desviación estándar global
print(M.var(axis=0)) # varianza de cada columna
```

```python
media = M.mean(axis=0, keepdims=True)
print(media.shape)
centrado = M - media
```
