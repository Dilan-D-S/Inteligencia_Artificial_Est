**Ejercicio 3:** Define una función 'resumen(lista)' que retorne un diccionario con: longitud, suma, promedio, mínimo y máximo de la lista.

```python
def resumen (lista):
  return {'longitud': len(lista),
          'suma': sum(lista),
          'promedio' : sum(lista)/len(lista),
          'minimo' : min(lista),
          'maximo' : max(lista)
          }

print(resumen([1, 2, 3, 4, 5]))
```

**Ejercicio 4:** Usa map() para aplicar la función potencia(x, 3) a la lista [1,2,3,4,5]

```python
def potencia(base,exponente=2):
  return base**exponente

lista = [1, 2, 3, 4, 5]

resultado = list(map(lambda x: potencia(x, 3), lista))
print(resultado)
```

# **Módulo 4 — Operaciones Numéricas y Manejo de Datos**

**Ejercicio 1:** Crea una matriz 5×4 de enteros aleatorios entre 1 y 100 (seed=7).


```python
np.random.seed(7)
A = np.random.randint(1, 101, (5, 4))

print("Matriz A:\n", A)
```

a) Calcula la media, máximo y mínimo de cada columna

```python
media = A.mean(axis=0)
maximo = A.max(axis=0)
minimo = A.min(axis=0)

print("Media:", media)
print("Máximo:", maximo)
print("Mínimo:", minimo)
```

b) Calcula la suma acumulada de la primera fila.


```python
acum = np.cumsum(A[0])
print("Suma acumulada (fila 1):", acum)
```

c) Normaliza la matriz para que cada columna tenga valores entre 0 y 1.
Fórmula: (x - min_col) / (max_col - min_col)


```python
A_norm = (A - minimo) / (maximo - minimo)
print("Matriz normalizada:\n", A_norm)
```

**Ejercicio 4:** Crea dos matrices 3×4 y concaténalas vertical y horizontalmente.
Luego divide el resultado vertical en 3 partes iguales con vsplit.

```python
A = np.arange(12).reshape(3, 4)
B = np.arange(12, 24).reshape(3, 4)

# Concatenación
vertical = np.vstack((A, B))
horizontal = np.hstack((A, B))

print("Vertical:\n", vertical)
print("\nHorizontal:\n", horizontal)

# Dividir en 3 partes iguales
dividido = np.vsplit(vertical, 3)

print("\nDivisiones:")
for i, d in enumerate(dividido):
    print(f"Parte {i}:\n", d)
```
